## **COVID 19**

La enfermedad por coronavirus (COVID-19) es una enfermedad infecciosa causada por un coronavirus recién descubierto. La mayoría de las personas infectadas con el virus COVID-19 experimentarán una enfermedad respiratoria leve a moderada y se recuperarán sin requerir tratamiento especial. El objetivo principal de este proyecto es construir un modelo de aprendizaje automático que, dado el síntoma, el estado y el historial médico actuales de un paciente con Covid-19, prediga si el paciente está en alto riesgo o no.

El conjunto de datos sin procesar consta de 21 características únicas y 1.048.576 pacientes únicos. En las características booleanas, 1 significa "sí" y 2 significa "no". valores como 97 y 99 faltan datos.

***Sexo:*** 1 para mujeres y 2 para hombres. 
Edad: del paciente. 

***Clasificación:*** Resultados de la prueba COVID. Los valores 1-3 significan que el paciente fue diagnosticado con covid en diferentes grados. 4 o superior significa que el paciente no es portador de covid o que la prueba no es concluyente. 

***Tipo de paciente:*** tipo de atención que el paciente recibió en la unidad. 1 para regresar a casa y 2 para hospitalización. 

***Neumonía:*** si el paciente ya tiene inflamación de los sacos de aire o no. ***Embarazo:*** si la paciente está embarazada o no. 

***Diabetes:*** si el paciente tiene diabetes o no. 

***EPOC: ***Indica si el paciente tiene enfermedad pulmonar obstructiva crónica o no. 

***Asma:*** si el paciente tiene asma o no. Inmsupr: si el paciente está inmunosuprimido o no. 

***Hipertensión:*** si el paciente tiene hipertensión o no. 

***Cardiovascular:*** si el paciente tiene una enfermedad relacionada con el corazón o los vasos sanguíneos. 

***Renal crónica:*** si el paciente tiene enfermedad renal crónica o no. 

***Otra enfermedad:*** si el paciente tiene otra enfermedad o no. 

***Obesidad:*** si el paciente es obeso o no. Tabaco: si el paciente es consumidor de tabaco. 

***usmr:*** Indica si el paciente trató unidades médicas de primer, segundo o tercer nivel. 

***unidad médica:*** tipo de institución del Sistema Nacional de Salud que prestó la atención. 

**Intubado:** si el paciente estaba conectado al ventilador. 

***UCI:*** Indica si el paciente había sido ingresado en una Unidad de Cuidados Intensivos. 

***Fecha de fallecimiento:*** Si el paciente falleció indique la fecha de fallecimiento, y 9999-99-99 en caso contrario.

### Importamos librerías de Pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from google.colab import files
import missingno as msn


%cd '/content/gdrive/MyDrive'
```

```python
data = pd.read_csv('fifa_eda_stats.csv', encoding = 'utf-8',sep=',',na_values='n/a')
data.info()
```

```python
indice = pd.Index(data['ID'])
jugadores = pd.DataFrame(data, columns = ['Name','Age','Nationality','Potential','Position','BallControl','SprintSpeed','Interceptions','Composure','Stamina','Preferred Foot','Release Clause'])
jugadores.index = indice
jugadores.dropna(axis = 0, how = 'any',inplace = True),
jugadores.info()
```

```python
%matplotlib inline
msn.matrix(jugadores)
```

Se analizan que los valores no se repitan

```python
jugadores.duplicated()
```

```python
jugadores.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
# Creando set de pruebas y set de entrenamiento
housing_with_id = jugadores.reset_index()
housing_with_id.head(10)
```

```python
Pais = jugadores['Nationality'].value_counts()
print(Pais)
print(type(Pais))
print(Pais.index)
```

```python
x2 = list(Pais.index)
y2 = Pais.values
print(x2)
```

```python
mpl.style.use('bmh') 
fig, ax = plt.subplots(figsize=(80,40))
ax.bar(x2,y2)
ax.set_title('Cantidad de jugadores por país')
ax.set_xlabel('Pais')
ax.set_xticklabels(x2, rotation = 60, ha="right") 
ax.set_ylabel('Cantidad')
```

Podremos observar la cantidad de jugadores por nacionalidades que tenemos
